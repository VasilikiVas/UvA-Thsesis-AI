```python
import torch
import os

def input_data_X(folder_path, feature_size):
  # Get the list of all .pt files in the folder
  file_list = [f for f in os.listdir(folder_path) if f.endswith('.pt')]

  # Initialize an empty tensor to store the data  
  data_tensor = torch.empty(len(file_list), feature_size)

  # Load each file into the tensor
  for i, file_name in enumerate(sorted(file_list)):
      file_path = os.path.join(folder_path, file_name)
      loaded_data = torch.load(file_path)
      data_tensor[i, :] = loaded_data

  # Return the final tensor
  return data_tensor
```

```python
import numpy as np

def input_data_y(folder_path, feature_size):
  # Get the list of all .pt files in the folder
  file_list = [f for f in os.listdir(folder_path) if f.endswith('.pt')]

  # Initialize an empty array to store the data  
  data_label = ["" for x in range(len(file_list))]

  # Load each file into the array
  for i, file_name in enumerate(sorted(file_list)):
      #data_label[i] = label_dict_inv[file_name.split('_')[0]]
      data_label[i] = label_dict_inv[file_name]
  return np.array(data_label)
```
